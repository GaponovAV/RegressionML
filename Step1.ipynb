```python
def get_name_empty_col(_data_):
    list_column=(list(_data_.columns))
    list_empty=[]
    for i in list_column:
        if _data_[i].isnull().sum()!=0:
            list_empty.append([i,_data_[i].isnull().sum()])
    return list_empty
```

```python
def input_empty(_data_,i):
    df = pd.DataFrame()
    df = _data_[_data_["Cluster"]==i]
    df["FAO"].fillna(df["FAO"].median(skipna=True),inplace=True)
    df["P_kg/ga"].fillna(df["P_kg/ga"].median(skipna=True),inplace=True)
    df["K_kg/ga"].fillna(df["K_kg/ga"].median(skipna=True),inplace=True)
    return df
```

```python
from sklearn.metrics import mean_absolute_error
def mean_absolute_scaled_error(y_test, predictions, y_train):
    e_t = y_test - predictions
    scale = mean_absolute_error(y_train[1:], y_train[:-1])
    return np.mean(np.abs(e_t / scale))
mean_absolute_scaled_error(y_test,predictions,y_train)
```

```python
coef_table = pd.DataFrame(columns=['Coef','Name Colum'])
def show_model_coef():
    global coef_table
    for i in range(len(model.coef_)):
        coef_table = coef_table.append({'Coef': np.round(abs(model.coef_[i]),5),'Name Colum': x_test.columns[i]},ignore_index=True )
        
    result = coef_table.sort_values(['Coef','Name Colum'], ascending=[1, 0])
    return result
show_model_coef()

```
